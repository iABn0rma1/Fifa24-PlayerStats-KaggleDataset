```python
# for data manipulation and analysis
import pandas as pd 

# for numerical computations and array operations
import numpy as np

# for basic plotting functionalities
import matplotlib as plt
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec  # Used for creating complex layouts

# for statistical data visualization based on Matplotlib
import seaborn as sns  # Provides themes and datasets for enhanced visualizations

data = pd.read_csv('player_stats.csv', 
                encoding='ISO-8859-1')
data.columns.values
```

```python
df = pd.DataFrame(data)
df
```

```python
# Univariate Analysis
sns.histplot(df['age'], bins=20, kde=True)
```

```python
# Bivariate Analysis
sns.scatterplot(x='sprint_speed', y='stamina', data=df)
```

```python
# Mulitvariate Analysis
plt.figure(figsize=(25, 20))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
threshold = 100  # Threshold for grouping countries into 'Other'
country_counts = df['country'].value_counts()
top_countries = country_counts[country_counts >= threshold].index
df['country_grouped'] = df['country'].apply(lambda x: x if x in top_countries else 'Other')
# Filter out rows for top countries and 'Other' category
filtered_df = df[df['country_grouped'].isin(top_countries) | (df['country_grouped'] == 'Other')]
# Calculate average skills for each country group
average_skills = filtered_df.groupby('country_grouped')[['ball_control', 'dribbling', 'marking', 'slide_tackle', 
                                                          'stand_tackle', 'aggression', 'reactions', 'att_position', 
                                                          'interceptions', 'vision', 'composure']].mean()

# Radar chart
labels = average_skills.columns
num_vars = len(labels)
angles = list(range(0, 360, int(360/num_vars)))
angles += angles[:1]
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.set_theta_offset(np.deg2rad(90))
ax.set_theta_direction(-1)

for i, country in enumerate(average_skills.index):
    values = average_skills.loc[country].values.flatten().tolist()
    values += values[:1]
    ax.plot(np.deg2rad(angles), values, label=country, linewidth=2)
    ax.fill(np.deg2rad(angles), values, alpha=0.25)

ax.set_xticks(np.deg2rad(angles[:-1]))
ax.set_xticklabels(labels)
ax.set_yticklabels([])
ax.legend(title='Country', bbox_to_anchor=(1.1, 1))
ax.set_title('Average Skills Comparison by Country Group')

plt.show()
```

```python
# Categorical Data Analysis
plt.figure(figsize=(10, 6))
sns.countplot(x='country_grouped', data=df, palette='viridis')
plt.title(f'Countries Grouped by Player Count (Top {threshold} Countries)')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```

```python
# # Outlier Detection
plt.figure(figsize=(8, 6))
sns.swarmplot(y='height', data=df, color='purple')
plt.title('Height Distribution with Outliers')
plt.ylabel('Height')
plt.show()
```
